# Pandas Plotting
![Plotting in Pandas](../assets/pandas_plotting.png)

```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_iris = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/main/pandas/tests/io/data/csv/iris.csv")
```

## Scatter Plot Matrix

```python
pd.plotting.scatter_matrix(df_iris.drop("Name", axis=1), figsize=(8, 8), alpha=0.7)
plt.savefig("scatter_plot.png")
plt.show()
```

## Andrews Plot

```python
plt.figure(figsize=(12, 7))
pd.plotting.andrews_curves(df_iris, 'Name')
plt.savefig("andrews_plot.png")
```

## Parallel Coordinates Plot

```python
plt.figure(figsize=(12, 7))
pd.plotting.parallel_coordinates(df_iris, class_column="Name")
plt.savefig("par_coord_plot.png")
```
